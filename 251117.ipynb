```python
#33
nums1 = {1,2,3,4,5}
nums2 = {4,5,6,7,8}

print("차집합:", nums1.difference(nums2))
print("대칭차집합:", nums1.symmetric_difference(nums2))
```

```python
#35
a = {100, 200, 300, 400, 500}
b = {400, 500, 600, 700, 800}

# intersection_update
a_inter = a.copy()
a_inter.intersection_update(b)
print("intersection_update:", a_inter)

# difference_update
a_diff = a.copy()
a_diff.difference_update(b)
print("difference_update:", a_diff)

# symmetric_difference_update
a_sym = a.copy()
a_sym.symmetric_difference_update(b)
print("symmetric_difference_update:", a_sym)
```

```python
#36
a = {100, 200, 300, 400, 500}
b = {100, 200, 300, 400, 500}

if a.issuperset(b) and a.issubset(b):
    print("동시")
elif a.issuperset(b):
    print("상위")
elif a.issubset(b):
    print("부분")
```
