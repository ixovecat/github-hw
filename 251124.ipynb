```python
#41
def get_max_min(a, b, c, d):
    scores = [a, b, c, d]
    return max(scores), min(scores)

a, b, c, d = map(int, input("점수 4개 입력: ").split())
max_score, min_score = get_max_min(a, b, c, d)
print("최대값:", max_score)
print("최소값:", min_score)
```

```python
#43
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

n = int(input("정수 n 입력: "))
print(factorial(n))
```

```python
#44
def sum_over_30(j):
    total = 0
    for i in range(2, 10):
        if i * j >= 30:
            total += i * j
    return total

j = int(input("양수 j 입력: "))
print(sum_over_30(j))
```

```python
#45
def cumulative_sum(lst):
    total = 0
    for x in lst:
        total += x
    return total

a = [1, 2, 3, 4, 5]
print(cumulative_sum(a))
```
